# Query the number of changes by status


```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
file = r'C:\Users\sddhr\Downloads\DA-WS-files\t_change.xlsx'
output = 'output.xlsx'
engine = create_engine('sqlite://',echo = False)
df = pd.read_excel(file) 
df.to_sql('change',engine,if_exists = 'replace',index = False)
results = engine.execute("SELECT COUNT(*), ch_status FROM change GROUP BY ch_status")
final = pd.DataFrame(results)
# final.to_excel(output,index = False)
final
```
