```python
import pandas as pd
import numpy as np

# importing the files

Input = pd.read_excel(r"I:\North America\USEH, East Hanover\PH\BusUnits\FFO\2. Targeting\2022C1\1. Oncology\1.Input\Tier files\STBL\KSQ_TIER_MOCK.XLSX",'Duplicates')

# 1st Qc: Valid all columns are present or not
Input.columns
#shape and size: 
# Input.shape

```

```python
# dropping unnecessary columns
import pandas as pd
df = pd.DataFrame(Input).drop(['FF_ID','START_DATE','END_DATE'],axis = 1)
df
```

```python
# Treating missing values
# There should not be missing values in any of the columns
# df.dropna(how = 'any').shape
df.notnull()
# count null values
df.isnull().sum()
```

```python
# FOR NGE BRANDS
df1=df[df['SEGMENT_TYPE']=="Segment"] 
df1

```

```python
df2=df[df['SEGMENT_TYPE']=="Tier"]
df2
```

```python
df3=df[df['SEGMENT_TYPE']=="RANK"]
df3
```

```python
# for  NGE BRANDS
df4 = df1.merge(df2,on = ['CUSTOMER_ID','PRODUCT_ID'])\
.merge(df3,on =['CUSTOMER_ID','PRODUCT_ID']).drop(['SEGMENT_TYPE_x','SEGMENT_TYPE_y','SEGMENT_TYPE'],axis =1)
df4
```

```python
# FOR NON NGE BRANDS
df4 = df2.merge(df3,on = ['CUSTOMER_ID','PRODUCT_ID']).drop(['SEGMENT_TYPE_x','SEGMENT_TYPE_y'],axis =1)
df4
```

```python
# FOR NGE BRANDS
df4.rename(columns = {'CUSTOMER_ID':'MDM_ID','PRODUCT_ID_x':'PROD','SEGMENT_VALUE_x':'SEGMENT','SEGMENT_VALUE_y':'TIER','SEGMENT_VALUE':'RANK'},inplace = True)
df4
```

```python
# FOR NON NGE BRANDS
df4.rename(columns = {'CUSTOMER_ID':'MDM_ID','PRODUCT_ID_x':'PROD','SEGMENT_VALUE_x':'TIER','SEGMENT_VALUE_y':'RANK'},inplace = True)
df4
```

```python
# duplicate tiers
duplicate_tier = df4[df4.duplicated(['MDM_ID','TIER'])]
print("Duplicate Rows based on MDM_ID & TIER :")
  
# Print the resultant Dataframe
duplicate_tier
```

```python
# duplicate ranks
duplicate_rank = df4[df4.duplicated(['MDM_ID'])]
print("Duplicate Rows based on MDM_ID & RANK :")
duplicate_rank
```

```python
# Count of records
df4.groupby(['TIER','SEGMENT']).size()
```

```python
# length check
df4["MDM_ID"]= df4["MDM_ID"].astype(str)
  
# passing values
df4["LENGTH"]= df4["MDM_ID"].str.len()

```

```python
# Check any other LENGTH value for MDM ID
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
# file = r'C:\Users\sddhr\Downloads\DA-WS-files\t_change.xlsx'
# output = 'output.xlsx'
engine = create_engine('sqlite://',echo = False)
# df = pd.read_excel(file) 
df4.to_sql('Input',engine,if_exists = 'replace',index = False)
results = engine.execute("SELECT distinct LENGTH FROM Input")
final = pd.DataFrame(results)
# final.to_excel(output,index = False)
final
```

```python
# LOWER TIER SHOULD HAVE BETTER RANKS
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
# file = r'C:\Users\sddhr\Downloads\DA-WS-files\t_change.xlsx'
# output = 'output.xlsx'
engine = create_engine('sqlite://',echo = False)
# df = pd.read_excel(file) 
df4.to_sql('Input',engine,if_exists = 'replace',index = False)
results = engine.execute("SELECT distinct TIER AS PROD_TIER,Max(RANK) AS MAXIMUM,MIN(RANK) AS MINIMUM FROM Input GROUP BY TIER ")
final = pd.DataFrame(results)
# final.to_excel(output,index = False)
final
```
